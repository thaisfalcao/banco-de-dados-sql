```python
i, j = 'Product ID', 'Product Name'

def check_unique(table, code_1, code_2):
    col = [code_1, code_2]
    aux = df[col]
    codes = list(df[code_1].drop_duplicates())
    cont = 0
    for code in codes:
        if len(aux[aux[code_1] == code].drop_duplicates()) > 1:
            cont += 1

    table.loc[code_1, code_2] = cont
    return table

cols = df.columns
df_num_repeat = pd.DataFrame(np.nan, index = cols, columns = cols)
df_num_repeat

# check_unique(df_num_repeat, i, j)
```

# Normalização
### Primeiro, escolhe-se as separações baseadas em relações de 1:1

```python
# Colunas de df
separations = [
    ['Postal Code', 'State', 'Region', 'Country'],
    ['Customer ID', 'Customer Name', 'Segment'],
    ['Order ID'],
    ['Product Name', 'Product ID', 'Sub-Category', 'Category'],
    ['Sales'],
    ['Order ID', 'Order Date', 'Ship Date', 'Ship Mode', 'City'],
 ]

# Códigos para chave Foreign
name_sep = [
    'cod_address',
    'cod_customer',
    'cod_id',
    'cod_prod',
    'cod_sales',
    'cod_ship',
]

list_of_codes = [
    'cod_address', 
    'cod_customer', 
    'cod_ship', 
    'cod_prod', 
    'cod_sales'
]
```

```python
path1 = 'Ada\\Dataset_Ada\\Tables1\\'

# Cria a função que cria uma tabela que normaliza o banco de dados. É importante que columns[0] seja o termo organizador do banco de dados.

def create_db(table, columns, code_name):
    df = table[columns].drop_duplicates().copy()
    num_line_aux = pd.factorize(df.apply(tuple, axis=1)) # Lembrar que Factorize não funciona com list
    df[code_name] = num_line_aux[0] + 1
    df.index = df[code_name]
    df = df[columns]
    return df
    
```

```python
create_db(df, separations[1], name_sep[1])
```

```python
path = os.getcwd() + '\\Tables\\'
os.listdir(path)
tables = [path + i for i in os.listdir(path)]
```

```python
list_of_tables = []

for col, code in zip(separations, name_sep):
    print(col, code)
    clear_output(wait = True)
    list_of_tables.append(create_db(df, col, code))
```

```python
list_of_tables[5]
```

# Criamos a tabela normalizada.


```python
df_norm = df.copy()
for table, columns in zip(list_of_tables, separations):
    df_norm = df_norm.merge(table.reset_index(), on = columns)
df_norm[name_sep]
```

```python
df_norm.to_csv(f'{path}main.csv')
for table, name_table in zip(list_of_tables, name_sep):
    table.to_csv(f'{path}{name_table}')
```

# Perguntas

1) Quais foram os 10 clientes que realizaram mais pedidos ao longo do período analisado?

```python
query = """
SELECT
    c.customer_name AS clientes,
    COUNT(c.customer_name) AS pedidos
FROM projeto_ada.main AS m
    LEFT JOIN projeto_ada.cod_customer AS c ON m.cod_customer=c.cod_customer
GROUP BY c.customer_name
ORDER BY COUNT(c.customer_name) DESC
LIMIT 10
;
"""

df = pd.read_sql_query(query, con = conn)
df
```

2) Quais foram os 10 clientes que mais gastaram dinheiro em compras ao longo do período analisado?

```python
query = """
SELECT
    c.customer_name AS clientes,
    ROUND(CAST(SUM(s.sales) AS NUMERIC), 2) AS vendas
FROM projeto_ada.main AS m
    LEFT JOIN projeto_ada.cod_customer AS c ON m.cod_customer=c.cod_customer
    LEFT JOIN projeto_ada.cod_sales AS s ON m.cod_sales=s.cod_sales
GROUP BY c.customer_name
ORDER BY SUM(s.sales) DESC
LIMIT 10
;
"""

df = pd.read_sql_query(query, con = conn)
df
```

3) Quais foram os 10 produtos mais vendidos durante o período de vendas analisado?

```python
query = """
SELECT
    p.product_name AS produtos,
    COUNT(p.product_name) AS pedidos
FROM projeto_ada.main AS m
    LEFT JOIN projeto_ada.cod_prod AS p ON m.cod_prod=p.cod_prod
GROUP BY p.product_name
ORDER BY COUNT(p.product_name) DESC
LIMIT 10
;
"""

df = pd.read_sql_query(query, con = conn)
df
```

4) Quais foram os 5 estados para os quais mais foram enviados pedidos no período analisado?

```python
query = """
SELECT
    a.state AS estados,
    COUNT(a.state) AS pedidos
FROM projeto_ada.main AS m
    LEFT JOIN projeto_ada.cod_address AS a ON m.cod_address=a.cod_address
GROUP BY a.state
ORDER BY COUNT(a.state) DESC
LIMIT 5
;
"""

df = pd.read_sql_query(query, con = conn)
df
```

5) Quantos pedidos cada região do EUA realizou?

```python
query = """
SELECT
    a.region AS regiões,
    COUNT(a.region) AS pedidos
FROM projeto_ada.main AS m
    LEFT JOIN projeto_ada.cod_address AS a ON m.cod_address=a.cod_address
GROUP BY a.region
ORDER BY COUNT(a.region) DESC
;
"""

df = pd.read_sql_query(query, con = conn)
df
```

6) Quais foram as 5 sub-categorias mais vendidas durante o período analisado?

```python
query = """
SELECT
    p.sub_category AS sub_categorias,
    ROUND(CAST(SUM(s.sales) AS NUMERIC), 2) AS lucro
FROM projeto_ada.main AS m
    LEFT JOIN projeto_ada.cod_prod AS p ON m.cod_prod=p.cod_prod
    LEFT JOIN projeto_ada.cod_sales AS s ON m.cod_sales=s.cod_sales
GROUP BY p.sub_category
ORDER BY SUM(s.sales) DESC
LIMIT 5
;
"""

df = pd.read_sql_query(query, con = conn)
df
```

7) Qual foi o ticket médio de cada categoria no período de vendas analisado?

```python
query = """
SELECT
    p.category AS categorias,
    ROUND(CAST(AVG(s.sales) AS NUMERIC), 2) AS ticket_medio
FROM projeto_ada.main AS m
    LEFT JOIN projeto_ada.cod_prod AS p ON m.cod_prod=p.cod_prod
    LEFT JOIN projeto_ada.cod_sales AS s ON m.cod_sales=s.cod_sales
GROUP BY p.category
ORDER BY AVG(s.sales) DESC
;
"""

df = pd.read_sql_query(query, con = conn)
df
```

8) Quantos pedidos foram realizados por cada tipo de cliente no período de vendas analisado?

```python
query = """
SELECT
    c.segment AS segmentos,
    COUNT(c.segment) AS pedidos
FROM projeto_ada.main AS m
    LEFT JOIN projeto_ada.cod_customer AS c ON m.cod_customer=c.cod_customer
GROUP BY c.segment
ORDER BY COUNT(c.segment) DESC
;
"""

df = pd.read_sql_query(query, con = conn)
df
```

9) Quantas vendas foram realizadas em cada ano do período analisado?

```python
query = """
SELECT
    DATE_PART('year', order_date) AS ano,
    COUNT(DATE_PART('year', order_date)) AS pedidos
FROM projeto_ada.main AS m
    LEFT JOIN projeto_ada.cod_ship AS s ON m.cod_ship=s.cod_ship
GROUP BY ano
ORDER BY ano
;
"""

df = pd.read_sql_query(query, con = conn)
df
```

10) Qual foi o tempo médio entre o pedido e a postagem para cada região do país no período de vendas analisado?

```python
query = """
SELECT
    a.region AS regiões,
    AVG(s.ship_date - s.order_date) AS dias
FROM projeto_ada.main AS m
    LEFT JOIN projeto_ada.cod_ship AS s ON m.cod_ship=s.cod_ship
    LEFT JOIN projeto_ada.cod_address AS a ON m.cod_address=a.cod_address
GROUP BY a.region
ORDER BY dias DESC
;
"""

df = pd.read_sql_query(query, con = conn)
df
```
